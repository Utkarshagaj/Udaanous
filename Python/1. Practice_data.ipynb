```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('1. practice_data.csv')
```

```python
df = pd.DataFrame(data)
df.sample()
```

```python
numeric_columns = ['Days Present', 'Sales Revenue Generated', ]
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for index, column in enumerate(numeric_columns):
    sns.kdeplot(data[column], ax=axes[index], fill=True)
    axes[index].set_title("KDE Plot")
plt.tight_layout()
plt.show()
```

```python
numeric_columns = ['Days Present', 'Sales Revenue Generated', ]
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for index, column in enumerate(numeric_columns):
    sns.kdeplot(df[column], ax=axes[index], fill=True)
    axes[index].set_title("KDE Plot")
plt.tight_layout()
plt.show()
```

```python
# Step 1: Check max sales revenue
max_revenue = df['Sales Revenue Generated'].max()
max_revenue
```

```python
# Step 2: Create a threshold — e.g., top 10% of revenue
threshold = df['Sales Revenue Generated'].quantile(0.90)
threshold
```

```python
# Step 3: Filter for the top 10% revenue values
top_revenue = df[df['Sales Revenue Generated'] >= threshold]
top_revenue
```

```python
# Step 4: Get the range
range_highest = top_revenue['Sales Revenue Generated'].min(), top_revenue['Sales Revenue Generated'].max()

print("Range of highest number of sales revenue generated:", range_highest)
```

```python
#Another method to calculate the overall range of number of sales revenue generated

revenue_range = df['Sales Revenue Generated'].min(), df['Sales Revenue Generated'].max()
print("Overall sales revenue range:", revenue_range)
```
